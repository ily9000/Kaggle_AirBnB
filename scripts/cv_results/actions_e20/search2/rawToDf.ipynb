```python
import pandas as pd
```

```python
res = pd.read_csv('res1.dat', delim_whitespace=True, header = None)
```

```python
df = pd.DataFrame()
for i in range(6):
    df = pd.concat([df, res.iloc[0+(i*50):(i+1)*50].reset_index()], axis = 1, ignore_index= True)
```

```python
df.columns[[0,1,4,5,8,9,12,13,16,17,20,21]]
```

```python
df.drop(df.columns[[0,1,4,5,8,9,12,13,16,17,20,21]], axis = 1, inplace = True)
```

```python
df = df.applymap(lambda x: x.lstrip('train-error:'))
```

```python
df = df.applymap(lambda x: x.lstrip('val-error:'))
```

```python
for i,j in enumerate(range(2,23,4)):
    df.rename(columns={j: 'train_error'+'_'+str(i)}, inplace = True)
```

```python
for i,j in enumerate(range(3,24,4)):
    df.rename(columns={j: 'eval_error'+'_'+str(i)}, inplace = True)
```
