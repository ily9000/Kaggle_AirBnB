```python
#%matplotlib inline
import numpy as np
import pandas as pd
import scipy.sparse
from sklearn import preprocessing
#import matplotlib.pyplot as plt
import xgboost as xgb
#import graphviz
```

Read in just the testing and training data.

```python
#train_100 = pd.read_csv('Airbnb_data/train_users_2.csv', nrows = 100, index_col='id')
train_full = pd.read_csv('data/train_users_2.csv', index_col = 'id')
test = pd.read_csv('data/test_users.csv', index_col = 'id')
#sessions = pd.read_csv('Airbnb_data/sessions.csv')
#age = pd.read_csv('Airbnb_data/age_gender_bkts.csv')
#countries = pd.read_csv('Airbnb_data/countries.csv')
```

Merge training and test data together.
In the future, change the session_flow to a categorical value.

```python
merge_data = pd.concat([train_full, test], axis = 0, join = 'inner')
print 'Are all columns of the test dataframe in the concatenated data frame?', all(merge_data.columns == test.columns)
print merge_data.dtypes 
```

```python
#Data in each column?
for col in merge_data:
#    if merge_data[col].dtypes == 'object':
#        print col, len(merge_data[col].unique()), merge_data[col].unique()
    print col, pd.isnull(merge_data[col]).values.any()
```

```python
#10 bins: greater than 1000
merge_data.loc[merge_data.query('age<18').index,'missing_age'] = 1
merge_data.loc[merge_data.query('18<=age<25').index,'18to25'] = 1
merge_data.loc[merge_data.query('25<=age<30').index,'25to30'] = 1
merge_data.loc[merge_data.query('30<=age<40').index,'30to40'] = 1
merge_data.loc[merge_data.query('40<=age<50').index,'40to50'] = 1
merge_data.loc[merge_data.query('50<=age<60').index, '50to60'] = 1
merge_data.loc[merge_data.query('60<=age<70').index, '60to70'] = 1
merge_data.loc[merge_data.query('70<=age<80').index, '70to80'] = 1
merge_data.loc[merge_data.query('80<=age<100').index, '80to100'] = 1
merge_data.loc[merge_data.query('100<=age<1000').index,'age_100'] = 1
merge_data.loc[merge_data.query('age>=1000').index,'age_1000'] = 1
merge_data.loc[pd.isnull(merge_data.age),'missing_age'] = 1
merge_data.iloc[:,14:25] = merge_data.iloc[:,14:25].fillna(0)
```

```python
#####Feature engineering#######
#drop date of first booking since that value is not found in the test data
merge_data.drop('date_first_booking', axis = 1, inplace = True)
#merge_data.drop('age', axis = 1, inplace=True)
merge_data = merge_data.fillna(-1)

#date_account_created
dac = np.vstack(merge_data.date_account_created.astype(str).apply(lambda x: list(map(int, x.split('-')))).values)
merge_data['dac_year'] = dac[:,0]
merge_data['dac_month'] = dac[:,1]
merge_data['dac_day'] = dac[:,2]
#merge_data['dac_dayofweek'] = dac[:,2]
merge_data.drop('date_account_created', axis=1, inplace = True)

#timestamp_first_active
tfa = np.vstack(merge_data.timestamp_first_active.astype(str).apply(lambda x: list(map(int, [x[:4],x[4:6],x[6:8],x[8:10],x[10:12],x[12:14]]))).values)
merge_data['tfa_year'] = tfa[:,0]
merge_data['tfa_month'] = tfa[:,1]
merge_data['tfa_day'] = tfa[:,2]
#merge_data['tfa_dayofweek'] = merge_data['timestamp_first_active'].astype(str)
merge_data.drop('timestamp_first_active', axis=1, inplace = True)

#age
av = merge_data.age.values
merge_data['age'] = np.where(np.logical_or(av<14, av>100), -1, av)
```

```python
#one hot encoding of features
print 'number of columns before one hot encoding', merge_data.shape[1]
ohe_feats = ['gender', 'signup_method', 'signup_flow', 'language', 'affiliate_channel', 'affiliate_provider', 'first_affiliate_tracked', 'signup_app', 'first_device_type', 'first_browser']
merge_data = pd.get_dummies(merge_data, columns = ohe_feats)
print 'number of columns after one hot encoding', merge_data.shape[1]

#convert targets to numerical values
targets = train_full['country_destination']
targets_le = preprocessing.LabelEncoder()
targets = targets_le.fit_transform(targets)
#convert targets in validation set
#validation_Y = targets_le.transform(validation_Y)
```

```python
#create a validation set
def get_validation(all_df, train_idx, trainY, n = 4000):
    train_df = all_df.loc[train_idx, :]
    train_df['country_destination'] = trainY
    valid_df = train_df.sample(4000)
    train_df.drop(valid_df.index, inplace = True)
    tr_X = train_df.iloc[:, 0:-1].values
    tr_Y = train_df.iloc[:, -1].values
    valid_X = valid_df.iloc[:, 0:-1].values
    valid_Y = valid_df.iloc[:, -1].values
    return tr_X, tr_Y, valid_X, valid_Y

train_X, train_Y, valid_X, valid_Y = get_validation(merge_data, train_full.index, targets)
```

```python
#set up data for xgboost
test_X = merge_data.ix[test.index, :].values

xg_train = xgb.DMatrix(train_X, label = train_Y, missing = -1)
xg_test = xgb.DMatrix(test_X, missing = -1)
xg_valid = xgb.DMatrix(valid_X, label= valid_Y, missing = -1)
```
